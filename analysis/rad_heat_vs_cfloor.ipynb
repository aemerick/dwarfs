```python
%matplotlib inline

import yt
import numpy as np
import matplotlib.pyplot as plt
```

```python
LT_dir = '/home/emerick/Research/dwarfs/flash_runs/leo_T/SN_LT_n150_v4_nh4/'
sim_dirs = {'low_pe':LT_dir + '0000_noUV_low-pe_nsn/',
            'heat':LT_dir + '0000_heat_nsn/',
            'no_UV':LT_dir + '0000_noUV_nsn/',
            'cfloor' : LT_dir + '0000_cfloor_nsn/'}
```

```python
data_dict = {}
for name in sim_dirs:
    print name
    data_dict[name] = np.genfromtxt(sim_dirs[name] + 'grav_bound_mass.dat',names=True)
```

```python
colors = {'low_pe':'orange','heat':'black','no_UV':'red','cfloor':'blue'}

for name in data_dict:
    m = data_dict[name]['m']
    plt.plot(data_dict[name]['t'], m/m[0], color=colors[name], lw=2.5, label=name)

plt.legend(loc='best')
```
